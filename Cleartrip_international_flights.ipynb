```python
airline_number=[]
airline_name=[]
Airline_Number=[]
Airline_Name=[]
for i in Airline:
    res=requests.get(i)
    soup=BeautifulSoup(res.text,'html.parser')
    r=soup.find_all('span',{'class':'name mt-15'})
    r1=soup.find_all('span',{'class':'name mt-5'})
    
    for i,j in zip(r,r1):
        #print(i.text,j.text)
        airline_number.append(i.text)
        airline_name.append(j.text)
    
for i,j in zip(airline_number,airline_name):
    i=i.split('\n')
    j=j.split('\n')
    Airline_Number.append(i[1].strip())
    Airline_Name.append(j[1])    
```

```python
import csv
with open('airline.csv','w',newline='') as f:
    head=['Airline_Name','Airline_Number']
    writer=csv.DictWriter(f,fieldnames=head)
    writer.writeheader()
    for i,j in zip(Airline_Name,Airline_Number):
        writer.writerow({'Airline_Name':i,'Airline_Number':j})
```
